Split the dataset into X and Y based on below list:
x = 'Mapped_Item_Code', 'Median_Price', 'Average_Tax_Percent',  'Effective_Price', 'Percent_Quantity_With_Discount',       'Avg_Discount_Percent_On_Discounted_Items','Car_sales', 'F_and_B_Inflation',       'F_and_B_Inflation_Adjusted_Effective_Price',       'F_and_B_Inflation_Adjusted_Median_Price'.
And  y = ‘Quantity_Sold’


```python
x=df_read1[['Mapped_Item_Code', 'Median_Price', 'Average_Tax_Percent', 'Effective_Price', 'Percent_Quantity_With_Discount', 'Avg_Discount_Percent_On_Discounted_Items','Car_sales', 'F_and_B_Inflation', 'F_and_B_Inflation_Adjusted_Effective_Price', 'F_and_B_Inflation_Adjusted_Median_Price']]
```

```python
y=df_read1[['Quantity_Sold']]
```

```python
x.head()
```

```python
y[:6]
```

Run the Linear Regression model on the above x and y features. 

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 75)
```

Creating an object for the LinearRegression function

```python
lr = LinearRegression()
```

Fitting into the model

```python
lr.fit(x_train, y_train)
```

Predicting the Output for the test set

```python
y_train_pred = lr.predict(x_test)
```

Finding the mean squared error for the test set 

```python
metrics.mean_squared_error(y_test,y_train_pred)
```

R-square of the model fit

```python
print("R-squared value =",round(metrics.r2_score(y_train_pred,y_test),3))
```

```python
print("R-squared value =",round(metrics.r2_score(y_test,y_train_pred),3))
```
